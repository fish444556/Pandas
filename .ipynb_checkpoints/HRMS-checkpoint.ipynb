## 1. Select First name, Last name, Address, Country Name (by joining with country table), Sate Name (by joining with state table), City Name (by joining with city table),Mobile, and Designation name (by joining with designation table) of all employees .

```python
df_q1 = pd.merge(df_emp, df_country, how="left", left_on='Emp_Country_Id', right_on='Country_Id')
df_q1 = pd.merge(df_q1, df_state, how="left", left_on='Emp_State_Id', right_on='State_Id')
df_q1 = pd.merge(df_q1, df_city, how="left", left_on='Emp_City_Id', right_on='City_Id')
df_q1 = pd.merge(df_q1, df_designation, how="left", left_on='Desig_Id', right_on='Desig_Id')
df_q1[['Emp_First_Name', 'Emp_Last_Name', 'Emp_Address1', 'Emp_Address2', 'Country_Name', 'State_Name', 'City_Name', 'Emp_Mobile', 'Designation']]
```

## 16. Fetch First name, last name, middle name, Country name, Designation name and DOB within a range (e.g.: 1/2/1984 to 1/3/1986).

```python
df_q16 = pd.merge(df_country, df_emp, how = 'right', left_on = 'Country_Id', right_on = 'Emp_Country_Id')
df_q16 = df_q16[(df_q16.Emp_DOB > '1984-02-01') & (df_q16.Emp_DOB < '1986-11-03')]
df_q16 = pd.merge(df_q16, df_designation, how = 'left', left_on = 'Desig_Id', right_on = 'Desig_Id')
df_q16[['Emp_First_Name', 'Emp_Last_Name', 'Emp_Middle_Name', 'Country_Name', 'Designation', 'Emp_DOB']]
```

## 17. Fetch First name, Last name, middle name, Country name, Designation name and highest salary.

```python
highest_idx = df_salary.Emp_Salary.idxmax()
df_q17 = pd.DataFrame({'Emp_Id': [df_salary.iloc[highest_idx].Emp_Id], 'Emp_Salary': [df_salary.iloc[highest_idx].Emp_Salary]})
df_q17 = pd.merge(df_q17, df_emp, how = 'left', left_on = 'Emp_Id', right_on = 'Emp_Id')
df_q17 = pd.merge(df_q17, df_country, how = 'left', left_on = 'Emp_Country_Id', right_on = 'Country_Id')
df_q17 = pd.merge(df_q17, df_designation, how = 'left', left_on = 'Desig_Id', right_on = 'Desig_Id')
df_q17[['Emp_First_Name', 'Emp_Last_Name', 'Emp_Middle_Name', 'Country_Name', 'Designation', 'Emp_Salary']]
```

## 18. Fetch First name, Last name, middle name, Country name, Designation name and his current salary.

```python
df_q18 = df_salary[df_salary.groupby('Emp_Id')['Emp_Salary_Change_Year'].transform('max') == df_salary['Emp_Salary_Change_Year']]
df_q18 = pd.merge(df_emp, df_q18, how="right", left_on='Emp_Id', right_on='Emp_Id')
df_q18 = pd.merge(df_q18, df_country, how = 'left', left_on = 'Emp_Country_Id', right_on = 'Country_Id')
df_q18 = pd.merge(df_q18, df_designation, how = 'left', left_on = 'Desig_Id', right_on = 'Desig_Id')
df_q18[['Emp_First_Name', 'Emp_Last_Name', 'Emp_Middle_Name', 'Country_Name', 'Designation', 'Emp_Salary']]
```

## 19. Fetch First name, Last name, middle name, Country name, Designation name and his/her highest salary within a range (e.g. : 50000 to 100000).

```python
df_salary_5k_10k = df_salary[~df_salary.Emp_Id.isin(df_salary[(df_salary['Emp_Salary'] < 50000) | (df_salary['Emp_Salary'] > 100000)]['Emp_Id'])]
df_q19 = df_salary_5k_10k.groupby('Emp_Id')['Emp_Salary'].max()
df_idx = pd.DataFrame({'Emp_Id': df_q19.keys(), 'Emp_Salary': df_q19.get_values() })
```

```python
df_designation.columns
```

```python
df_q19 = pd.merge(df_idx, df_emp, how='left', left_on='Emp_Id', right_on='Emp_Id')
df_q19 = pd.merge(df_q19, df_country, how='left', left_on='Emp_Country_Id', right_on='Country_Id')
df_q19 = pd.merge(df_q19, df_designation, how='left', left_on='Desig_Id', right_on='Desig_Id')
df_q19[['Emp_First_Name', 'Emp_Last_Name', 'Emp_Middle_Name', 'Country_Name', 'Designation', 'Emp_Salary']]
```

## 25. Get Designation, total salary spend by the company with respect to each E.g.

```python
salary_sum = df_salary.groupby('Emp_Id')['Emp_Salary'].sum()
df_salary_sum = pd.DataFrame({'Emp_Id': salary_sum.keys(), 'Salary_Sum': salary_sum.get_values()})
df_q25 = pd.merge(df_emp, df_salary_sum, how = 'right', left_on = 'Emp_Id', right_on = 'Emp_Id')
df_q25 = pd.merge(df_designation, df_q25, how = 'right', left_on = 'Desig_Id', right_on = 'Desig_Id')
df_q25.groupby('Desig_Description')['Salary_Sum'].sum()
```

## 26. Get designation name, no of employees in a designation with respect to a designation from employee details table.

```python
df_q26 = pd.merge(df_designation, df_emp, how = 'right', left_on = 'Desig_Id', right_on = 'Desig_Id')
df_q26.groupby('Designation')['Emp_Id'].count()
```

## 27. Get designation wise average salaries from employee details table and order by Salary ascending.

```python
series_q27 = df_salary.groupby('Emp_Id')['Emp_Salary'].mean()
df_q27 = pd.DataFrame({'Emp_Id': series_q27.keys(), 'Salary_Sum': series_q27.get_values()})
```

```python
df_q27 = pd.merge(df_q27, df_emp, how='right', left_on='Emp_Id', right_on='Emp_Id')
df_q27 = pd.merge(df_q27, df_designation, how='left', left_on='Desig_Id', right_on='Desig_Id')
df_q27.groupby('Desig_Description')['Salary_Sum'].mean().sort_values(ascending=False)
```

## 30. Select first name, sum of salary from employee details and salary table for all employees even if they didn't get salary and set salary amount as 0 for those employees who didn't get salary.

## 32. Display how many days each employee worked in the company.

## 33. Select the employee with highest appraisal. (difference of first salary and current salary)

```python
highest_appraisal = (df_salary.groupby('Emp_Id')['Emp_Salary'].max() - df_salary.groupby('Emp_Id')['Emp_Salary'].min()).idxmax()
df_emp[df_emp['Emp_Id'] == highest_appraisal]
```
